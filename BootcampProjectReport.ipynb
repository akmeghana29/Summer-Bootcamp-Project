```python
# Problem Statement/ Objective

The transportation sector is a significant contributor to greenhouse gas emissions.
Electric vehicles (EVs) offer a promising solution for reducing emissions and promoting sustainable transportation. 
    To accelerate EV adoption, we need to understand the current market landscape and identify key trends influencing growth.

This project aims to analyze historical EV registration data from Washington state to gain insights into the current state of EV penetration. 
                                                                                                                                                                                                                                 By leveraging data analytics, we will investigate factors that drive EV ownership and identify areas for strategic intervention.

Our analysis will address the following questions:

What is the current market penetration of EVs in Washington state?
How are EVs distributed across different geographic regions, vehicle types, and price points?
What are the trends in EV registration by model year and electric range?
Are there any correlations between vehicle characteristics (e.g., range, price) and geographic location (e.g., county, legislative district)?
What insights can be gleaned from the data to inform future policies and infrastructure development for promoting wider EV adoption?
By answering these questions, this project will provide valuable information for stakeholders involved in the EV ecosystem, 
including policymakers, utility companies, and automotive manufacturers. The analysis will help guide strategies to accelerate 
EV adoption in Washington state and contribute to a cleaner transportation future.
```
