```python
# Converting some columns to numeric to avoid errors
data['Model Year'] = pd.to_numeric(data['Model Year'], errors='coerce')
data['Electric Range'] = pd.to_numeric(data['Electric Range'], errors='coerce')
data['Base MSRP'] = pd.to_numeric(data['Base MSRP'], errors='coerce')
```

```python
# 8. Check for anomalies or wrong entries
anomalies = data[(data['Model Year'] > 2024) | (data['Electric Range'] < 0) | (data['Base MSRP'] < 0)]
print("\nPotential anomalies or wrong entries:")
print(anomalies)
```

```python
# 9. Check the outliers using a more memory-efficient approach
outliers = pd.DataFrame()

for column in data.select_dtypes(include=[np.number]).columns:
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    column_outliers = data[(data[column] < (Q1 - 1.5 * IQR)) | (data[column] > (Q3 + 1.5 * IQR))]
    outliers = pd.concat([outliers, column_outliers])

outliers.drop_duplicates(inplace=True)
print("\nOutliers in the dataset:")
print(outliers)
```

```python
# 10. Data cleaning
# Drop duplicates
data_cleaned = data.drop_duplicates()

# Drop unnecessary columns (for example, VIN (1-10), DOL Vehicle ID)
columns_to_drop = ['VIN (1-10)', 'DOL Vehicle ID']
data_cleaned = data_cleaned.drop(columns=columns_to_drop, errors='ignore')

# Null value imputation
# For numerical columns, we can use the mean or median
# For categorical columns, we can use the mode
for column in data_cleaned.columns:
    if data_cleaned[column].isnull().sum() > 0:
        if pd.api.types.is_numeric_dtype(data_cleaned[column]):
            data_cleaned[column].fillna(data_cleaned[column].median(), inplace=True)
        else:
            data_cleaned[column].fillna(data_cleaned[column].mode()[0], inplace=True)

# Outlier treatment (for simplicity, we can remove them)
data_cleaned = data_cleaned[~data_cleaned.index.isin(outliers.index)]

print("\nData after cleaning:")
print(data_cleaned.head())
print("\nShape of cleaned dataset:", data_cleaned.shape)
```
